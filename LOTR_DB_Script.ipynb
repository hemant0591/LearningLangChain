Pushing Lord of the Rings - The fellowship of the rings data into vector store.

```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_postgres.vectorstores import PGVector
from langchain_openai import OpenAIEmbeddings
```

```python
raw_doc = TextLoader('data/FellowshipOfTheRing.txt', encoding='utf-8').load()
splitter = RecursiveCharacterTextSplitter(chunk_size=2000, chunk_overlap=200)
splitted_docs = splitter.split_documents(raw_doc)
```

```python
splitted_docs[0]
```

```python
# embed chunks and insert into the vector store

embeddings = OpenAIEmbeddings(model="text-embedding-3-small")
connection = 'postgresql+psycopg://langchain:langchain@localhost:6024/langchain'
collection_name = "Lord_of_the_rings"

db = PGVector.from_documents(
    documents=splitted_docs,
    embedding=embeddings,
    collection_name=collection_name,
    connection=connection
)
```

```python
db.similarity_search('Frodo Baggins', k=2)
```
